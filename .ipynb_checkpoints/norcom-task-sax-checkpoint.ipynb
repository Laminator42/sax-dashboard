```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(seed=42)
```

```python
def PAA(T, w):
    n = len(T)
    if n%w:
        # TODO: cut T if n is divisble by w
        raise ValueError("n not divisible by w.")
    t = [np.mean(T[n//w*(i-1)+1:n//w*i+1]) for i in range(1, w+1)]
    return t
        

xs = range(32)
ts = np.array([4*np.exp(0.05 * (x + np.random.normal(0, 3))) for x in xs])
```

```python
W = 8
plt.plot(range(len(ts)), ts)
plt.plot(range(len(ts)), np.repeat(PAA(ts, W), 32//W))
```
